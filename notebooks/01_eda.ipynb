```python
df = df.sort_values(['Product Name', 'Location', 'Platform', 'Date'])
```

```python
df['lag_1'] = df.groupby(['Product Name', 'Location', 'Platform'])['Units Sold'].shift(1)
df['lag_2'] = df.groupby(['Product Name', 'Location', 'Platform'])['Units Sold'].shift(2)
df['lag_4'] = df.groupby(['Product Name', 'Location', 'Platform'])['Units Sold'].shift(4)
```

```python
df.head()
```

```python
df['rolling_mean_4'] = (
    df.groupby(['Product Name', 'Location', 'Platform'])['Units Sold']
      .shift(1)
      .rolling(4)
      .mean()
)
```

```python
df.head()
```

```python
df['rolling_std_4'] = (
    df.groupby(['Product Name', 'Location', 'Platform'])['Units Sold']
      .shift(1)
      .rolling(4)
      .std()
)
```

```python
df.head()
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['week'] = df['Date'].dt.isocalendar().week.astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df['Date'].max()
```

```python
split_date = df['Date'].quantile(0.8)

train = df[df['Date'] <= split_date]
test = df[df['Date'] > split_date]
```

```python
train.shape, test.shape
```

```python
y_train = train['Units Sold']
y_test = test['Units Sold']
```

```python
features = [
    'lag_1', 'lag_2', 'lag_4',
    'rolling_mean_4', 'rolling_std_4',
    'year', 'month', 'week',
    'Price', 'Discount'
]

X_train = train[features]
X_test = test[features]
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
exclude_cols = ['Date', 'Units Sold', 'Revenue']
features = [col for col in df.columns if col not in exclude_cols]

split_date = df['Date'].quantile(0.8)

train = df[df['Date'] <= split_date]
test = df[df['Date'] > split_date]

X_train = train[features]
X_test = test[features]

y_train = train['Units Sold']
y_test = test['Units Sold']

```

```python
rf.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

y_pred = rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("MAE:", mae)
print("RMSE:", rmse)
print("MAPE:", mape)
```
